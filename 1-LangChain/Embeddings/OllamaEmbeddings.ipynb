#### Ollama
Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = OllamaEmbeddings(model="mxbai-embed-large")  ##by default it ues llama2
```

Embed Query/Text

```python
text = "This is a text document"
```

```python
query_results = embeddings.embed_query(text)
query_results
```

```python
len(query_results)
```

Embed Documents

```python
r1 = embeddings.embed_documents([
    "Alpha is the first letter of Greek alphabets",
    "Beta is the second letter of Greek alphabets",
])
```

```python
len(r1)
```

```python
len(r1[0])
```
